### F1. Shopping List Manager (1D list + functions) (Suggested)
Write a program that maintains a shopping list (a 1D list of strings).

**Required functions:**
- `add_item(items, item)`
- `remove_item(items, item)` (remove by value if present)
- `print_items(items)` (numbered list)
- `save_items(items, filename)` (one item per line)
- `load_items(filename)` (returns a list)

**Program behavior:**
Provide a simple text menu in a loop:
1) add  2) remove  3) view  4) save  5) load  6) quit
The list should update as the user makes choices.


````python
# TODO: Implement F1 here
shopping_list = []

# adds an item to the shopping list
def add_item(shopping_list, item):
  shopping_list.append(item)
  return shopping_list

# removes an item from the shopping list
def remove_item(shopping_list, item):

  if item in shopping_list:
    shopping_list.remove(item)

  else:
    return("That item is not in your list")
  return shopping_list

# prints the shopping list as a numbered list
def print_items(shopping_list):

  for index, item in enumerate(shopping_list):
    print(f"{index + 1}. {item}")

# saves the items to a file
def save_items(shopping_list, filename = "shopping.txt"):

  try:
    with open(filename, "w") as f:
      for food in shopping_list:
        food = food.strip()
        f.write(food + "\n")

  except FileNotFoundError:
    return("File not found")

  return shopping_list

# loads the file into a list
def load_items(filename = "shopping.txt"):

  loaded = []
  try:
    with open(filename, "r") as f:
      for line in f:
        loaded.append(line)

  except FileNotFoundError:
    return("File not found")

  return loaded, shopping_list

# Offers the menu and executes the functions
def main(shopping_list):

  # Prints the menu
  print("Menu: ")
  print("1: Add Item")
  print("2: Remove Item")
  print("3: Print Items")
  print("4: Save Items")
  print("5: Load Items")
  print("6: Quit")

  # Loops for user input
  while True:
    choice = input("Please select a choice from the menu    ")
    choice = int(choice)
    if choice not in range(1, 7):
      return("Please choose an option from the menu ")
    else:
      if choice == 1:
        item = input("Please enter an item to add  ")
        add_item(shopping_list, item)
      elif choice == 2:
        item = input("Please enter an item to remove  ")
        remove_item(shopping_list, item)
      elif choice == 3:
        print_items(shopping_list)
      elif choice == 4:
        save_items(shopping_list, filename = "shopping.txt")
      elif choice == 5:
        load_items(filename = "shopping.txt")
      else:
        print("Goodbye!")
        break

# calls the main function
main(shopping_list)
````

### F2. Grade Statistics (1D list of numbers + file input)
You will read a file of grades (one number per line) into a 1D list, then compute statistics.

**Requirements:**
- Read `grades.txt` into a list `grades`
- Compute and print: count, min, max, average
- Create a new list `curved` that adds +5 to each grade (cap at 100)
- Write `curved` to `curved_grades.txt` (one per line)

**If you don't have a `grades.txt` file, create one in code first** (write 8–12 sample grades).


````python
# TODO: Implement F2 here

grades = ["89", "98", "76", "96", "65", "100", "92", "80", "74", "93"]
curved = []

# writes the list of grades to a file
def create_file(grades):
  with open("grades.txt", "w") as f:
    for grade in grades:
      f.write(grade + "\n")

# calculates grade statistics
def F2():

  count = 0
  min = 101   # higher than the highest possible number
  max = -1    # lower than the lowest possible number
  total = 0

  try:
    with open("grades.txt", "r") as f:
      for line in f:
        number = int(line)    # changes the line fro a string to an integer
        count += 1
        total += number

        if number < min:
          min = number
        if number > max:
          max = number

    average = total/count
    print(f"The min is: {min} \n The max is: {max} \n The number of grades is: {count} \n The average is: {average}")

    curved_number = number + 5
    if curved_number > 100:
      curved_number = 100
    curved.append(curved_number)

    return curved

  except FileNotFoundError as e:
    return(f"Error: {e}")

# writes the curved grades to a file
def curved_grades(curved):

  try:
    with open("curved_grades.txt", "w") as f:
      for item in curved:
        f.write(str(item) + "\n")    # changes the line from an integer to a string

  except FileNotFoundError as e:
    return(f"Error: {e}")

# calls the functions
create_file(grades)
F2()
curved_grades(curved)
````

### F3. Word Cleaner (file → 1D list → processing → file)
Read a text file (e.g., `paragraph.txt`) and build a 1D list of cleaned words.

**Requirements:**
- Read the file as text
- Convert to lowercase
- Remove punctuation characters like `. , ! ? ; :` (you may replace them with spaces)
- Split into words and store in a list `words`
- Build a list `long_words` containing only words with length ≥ 6
- Write `long_words` to `long_words.txt`, one per line

**If you don't have a `paragraph.txt`, create one in code first** (write 2–4 sentences).


````python
# TODO: Implement F3 here

def F3():

  long_words = []

  # saves the words as a list
  with open("paragraph.txt", "r") as f:
    file = f.read()
    file = file.lower().replace(".", " ").replace(",", " ").replace("!", " ").replace("?", " ").replace(";", " ").replace(":", " ")
    words = file.split()

    # saves the long words to a list
    for word in words:
      if len(word) >= 6:
        long_words.append(word)

  # writes the long words to a new file
  with open("long_words.txt", "w") as f:
    for word in long_words:
      f.write(word + "\n")

F3()
````

### F4. Seating Chart (nested list) (2D list + loops + functions)
Create a seating chart using a nested list (list of rows).

**Requirements:**
- Build a 4×5 chart filled with the string `"EMPTY"`
- Write a function `assign_seat(chart, row, col, name)` that places a name
- Write a function `print_chart(chart)` that prints each row neatly
- Assign at least 6 names to different seats
- Count how many seats are still `"EMPTY"` and print that number

**Optional extension:** Write the seating chart to a file (one row per line).


````python
# TODO: Implement F4 here

# create empty seating chart
chart = [["EMPTY", "EMPTY", "EMPTY", "EMPTY"], ["EMPTY", "EMPTY", "EMPTY", "EMPTY"],
   ["EMPTY", "EMPTY", "EMPTY", "EMPTY"], ["EMPTY", "EMPTY", "EMPTY", "EMPTY"],
    ["EMPTY", "EMPTY", "EMPTY", "EMPTY"]]

# assign seats by indexing
def assign_seat(chart, row, col, name):

  row = row - 1
  col = col - 1
  chart[row][col] = name
  return chart

# print the chart neatly and calculate empty seats by running total
def print_chart(chart):

  count = 0
  for row in chart:
    print(f"{row[0]:>10} {row[1]:>10} {row[2]:>10} {row[3]:>10} \n")
    for item in row:
      if item == "EMPTY":
        count += 1
  print(f"Amount of empty seats: {count}")

# call the function with arguments to assign specific seats
assign_seat(chart, 3, 2, "Sarah")
assign_seat(chart, 1, 1, "Rivka")
assign_seat(chart, 2, 4, "Rochel")
assign_seat(chart, 5, 1, "Leah")
assign_seat(chart, 1, 2, "Yocheved")
assign_seat(chart, 2, 3, "Miriam")

# print the chart
print_chart(chart)


````
